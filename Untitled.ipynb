# Ans5

```python
# Levels of measurement, also called scales of measurement, . In scientific 
# research, a variable is anything that can take on different values 
# across your data set (e.g., height or test scores).
# Nominal: the data can only be categorized(Gender)
# Ordinal: the data can be categorized and ranked(Top 5 Olympic medallists)
# Interval: the data can be categorized, ranked, and evenly spaced(Temperature in Fahrenheit or Celsius)
# Ratio: the data can be categorized, ranked, evenly spaced, and has a natural zero.(Height)
```
